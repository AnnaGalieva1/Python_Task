```python
from sympy import *
from sympy import Float 
from sympy import Symbol
from sympy.plotting import plot 

print('Colab')
x = symbols('x')
f = -12*x**4*sin(cos(x))- 18*x**3+ 5*x**2+10*x-30
plot(f, (x, -100, 100), line_color='red')
dy = (diff(f))
print('Производная f(x)=',dy)
plot(dy, (x, -100, 100), line_color='violet')
#Зададим интервал (1.4, 2.5)
plot(f, (x, 1.4, 2.5), line_color='blue')
```
